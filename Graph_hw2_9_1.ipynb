```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
def create_G_t(t:int):
    G = nx.MultiDiGraph()
    G.add_node(0)
    G.add_node(1)
    G.add_edge(0, 1)
    if t!=0:
        list_degree=[]
        while G.number_of_nodes() < t+2:
            new_node = G.number_of_nodes()
            G.add_node(new_node)
            edge = np.random.choice(G.number_of_edges())#selected edge
            #endpoint of selected edge
            point1 = list(G.edges())[edge][0]
            point2 = list(G.edges())[edge][1]
            G.add_edge(new_node, point1)
            G.add_edge(new_node, point2)
            Sum_num_node=G.number_of_nodes()
            
            degree_num2 = 0
            for node in G.nodes():
                if G.degree(node) == 2:
                    degree_num2 = degree_num2 + 1
            list_degree.append(degree_num2 / Sum_num_node)
        return G, list_degree
    else: return G
```

```python
G1, list_degree_1 = create_G_t(t=2000)
G2, list_degree_2 = create_G_t(t=2000)
G3, list_degree_3 = create_G_t(t=2000)
G4, list_degree_4 = create_G_t(t=2000)
G5, list_degree_5 = create_G_t(t=2000)
G6, list_degree_6 = create_G_t(t=2000)
G7, list_degree_7 = create_G_t(t=2000)
plt.figure(figsize=(10,10))
plt.plot(list_degree_1)
plt.plot(list_degree_2)
plt.plot(list_degree_3)
plt.plot(list_degree_4)
plt.plot(list_degree_5)
plt.plot(list_degree_6)
plt.plot(list_degree_7)
plt.xlabel('range of t')
plt.ylabel('the proportion of points of degree 2 in the graph G_t')
plt.grid()
```
